```python
"""
Demo script for reading a CSV file from S3 into a pandas data frame using s3fs-supported pandas
APIs
"""

import os

import pandas as pd

AWS_S3_BUCKET = "ab2-348052051973"

key = "corpus/toxic_comments_train_preprocessed.csv"

toxic_df = pd.read_csv(
    f"s3://{AWS_S3_BUCKET}/{key}",
)

```

```python
df = toxic_df.copy()
df["label"] = ""
metrics = ["identity_hate", "insult", "obscene", "severe_toxic", "threat", "toxic"]
for metric in metrics:
    df.loc[df[metric] > 0.5, "label"] = df["label"] + metric+"|"

df.head(10)
```

```python
df["label"] = df["label"].str.strip("|")

df.loc[df["label"] == "", "label"] = "non-toxic"

df.head(10)
```

```python
export_df = df[["label", "comment_text"]]

export_file_name="s3://" + AWS_S3_BUCKET + "/corpus/toxic_train_set.csv"
export_df.to_csv(export_file_name, header=False, index=False )
```
